### Searching for related features in the cell data set

```python
#Import relevant libraries

import numpy as np

import pandas as pd

import pprint
pp = pprint.PrettyPrinter(indent=2, width=1)

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#Import cell type data

from allensdk.core.cell_types_cache import CellTypesCache
from allensdk.ephys.feature_extractor import EphysFeatureExtractor

ctc = CellTypesCache(manifest_file='D:/cell_types/manifest.json')

#specimen_id = 464212183
```

```python
# download ephys features for all cells
features = ctc.get_ephys_features()
df_ephys = pd.DataFrame.from_records(features)
df_ephys
```

```python
#Get morphology data for all cells
from allensdk.api.queries.rma_api import RmaApi

rma_api = RmaApi()
reconstructions = rma_api.model_query('NeuronReconstruction',num_rows='all')
```

```python
df_morph = pd.DataFrame.from_records(reconstructions)
df_morph
```

```python
len(set(df_morph.specimen_id))
```

```python
cells_with_morph = df_morph.specimen_id
len(cells_with_morph)
```

```python
# Collect cells with ephys and morphology data

pd.merge(df_ephys, df_morph, how='inner', on='specimen_id', left_on=None, right_on=None,
      left_index=False, right_index=False, sort=True,
      suffixes=('_x', '_y'), copy=True)
```
